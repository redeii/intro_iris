```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(random_state=42)
```

```python
from sklearn.svm import LinearSVC
```

```python
linear_svm = LinearSVC().fit(X,y)
```

```python
print('Coef Shape ',linear_svm.coef_.shape)
```

$y_i = w_i[0]*x_i[0] + w_i[1]*x_i[1] + b$

```python
%matplotlib inline
fig, ax = plt.subplots()
markers = ['o','^','v']
colors = ['blue','yellow','green']
for clss in set(y):
    ax.scatter(X[y==clss,0],X[y==clss,1],c=colors[clss],marker=markers[clss],edgecolors='black',label=f'Class {clss}',s=100)
line = np.linspace(-15,15)
for coef, intercept,color in zip(linear_svm.coef_,linear_svm.intercept_,colors):
    ax.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
ax.set_xlabel('Feature 0')
ax.set_ylabel('Feature 1')
ax.set_ylim(-10,15)
ax.set_xlim(-10,9)
ax.legend()
```

```python
line = np.linspace(-15,15)
for coef, intercept,color in zip(linear_svm.coef_,linear_svm.intercept_,['b','r','g']):
    plt.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
```
